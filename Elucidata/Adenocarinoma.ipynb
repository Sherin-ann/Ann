# Remove Neurocarcinomas from dataset 


```python
import pandas as pd
import csv
from cmapPy.pandasGEXpress.parse import parse
```


## Extract necessary columns from dataset

```python
sig_info = parse("PAAD.gct", col_meta_only=True) #parse the column metadata into sig_info
patient_id = sig_info["participant_id"]
patient_barcode= sig_info["bcr_patient_barcode"]
adc = sig_info["histological_type_other"]
adc1 = sig_info["histological_type"]
```

## Extract only adrenocarcinoma tumor samples from the dataset

```python
print("patient_id\tpatient_barcode\t\thistological_type\t\t\thistological_type_other") #print headers

for i in range(1,183):
    if ("adenocarcinoma" in str(adc[i])) or ("pancreas-adenocarcinoma-other subtype" not in str(adc1[i]) and "nan" not in str(adc1[i])):
        #print dataset containing only adenocaricoma tumor samples
        print(patient_id[i],"\t\t",patient_barcode[i],"\t\t",adc1[i],"\t\t",adc[i])
    
```
